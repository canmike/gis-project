```python
import pandas as pd
import numpy as np
```

```python
def process_air_pollution_df(df):
    # Rename first column
    df.rename(columns={'Unnamed: 0': 'Timestamp'}, inplace=True)
    
    # Convert timestamp
    df['Timestamp'] = pd.to_datetime(df['Timestamp'])
    
    # Process numeric columns
    numeric_cols = df.columns.drop('Timestamp')
    for col in numeric_cols:
        # Remove unit information from column names
        new_col = col.split('(')[0].strip()
        df.rename(columns={col: new_col}, inplace=True)
        
        try:
            # Handle string conversions
            df[new_col] = df[new_col].replace('-', np.nan)
            if df[new_col].dtype == 'object':
                df[new_col] = (df[new_col]
                              .str.replace('.', '', regex=False)
                              .str.replace(',', '.'))
            df[new_col] = pd.to_numeric(df[new_col], errors='coerce')
        except Exception as e:
            print(f"Error in column {new_col}:", e)
    
    return df
```

# 4. Gather All Traffic Density Data

```python
from tqdm import tqdm

# Define root path
root_path = 'trafficdensity_fixed'
output_path = 'all_traffic_density_data.csv'

# List all files in the directory
file_paths = os.listdir(root_path)

# Initialize an empty list to hold dataframes
data_frames = []

# Process each file
for file_name in tqdm(file_paths):
    file_path = os.path.join(root_path, file_name)
    if os.path.isfile(file_path) and file_name.endswith('.csv'):
        # Read the file
        df = pd.read_csv(file_path)
        # Check if the file has at least 8 columns
        if df.shape[1] >= 8:
            # Select the required columns (1, 2, 3, and 8)
            selected_columns = df.iloc[:, [0, 1, 2, 7]]
            selected_columns.columns = ['Timestamp', 'Latitude', 'Longitude', 'Vehicle Count']
            # Append to the list
            data_frames.append(selected_columns)
        else:
            print(f"Skipping {file_name}: Not enough columns.")

# Concatenate all dataframes into one
if data_frames:
    combined_df = pd.concat(data_frames, ignore_index=True)
    # Save the combined dataframe to a new CSV file
    combined_df.to_csv(output_path, index=False)
    print(f"Combined data saved to {output_path}")
else:
    print("No valid data to combine.")
```

## 8. Get Bounding Box of Istanbul

```python
import geopandas as gpd

# Use forward slashes or raw string for path
istanbul_shp_path = 'istanbul_sìnìr/isttt.shp'

istanbul_shp = gpd.read_file(istanbul_shp_path)
bbox = istanbul_shp.total_bounds
print(f"Bounding Box: {bbox}")
```

```python
import geopandas as gpd
from shapely.geometry import box
import os

def export_bbox(shp_path):
    gdf = gpd.read_file(shp_path)
    bbox = gdf.total_bounds
    bx = gpd.GeoDataFrame(
        geometry=[box(*bbox)],
        crs=gdf.crs
    )
    os.makedirs('istanbul_bbox', exist_ok=True)
    bx.to_file('istanbul_bbox/istanbul_bbox.shp')
    return bx

istanbul_shp_path = 'istanbul_sìnìr/isttt.shp'
bbox_gdf = export_bbox(istanbul_shp_path)
```

## 9. Convert Monthly Data to Shapefile

```python
monthly_air_pollution_folder = 'monthly_air_pollution_data'
monthly_air_pollution_files = os.listdir(monthly_air_pollution_folder)
```

```python
out_shp_folder = 'monthly_air_pollution_data_shp'
os.makedirs(out_shp_folder, exist_ok=True)

for file in monthly_air_pollution_files:
    file_path = os.path.join(monthly_air_pollution_folder, file)
    
    # Read CSV and create GeoDataFrame
    df = pd.read_csv(file_path)
    gdf = gpd.GeoDataFrame(
        df, 
        geometry=gpd.points_from_xy(df.Longitude, df.Latitude),
        crs="EPSG:4326"
    )
    
    # Spatial join with bbox
    gdf = gpd.sjoin(gdf, bbox_gdf, op='within')
    
    # Save to shapefile
    out_file = file.replace('.csv', '.shp')
    gdf.to_file(os.path.join(out_shp_folder, out_file))
```
